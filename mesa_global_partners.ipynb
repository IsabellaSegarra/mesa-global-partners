# Multinational Exchange for Sustainable Agriculture Global Partners 

```python
# Library set up
import pandas as pd # for data wrangling
import numpy as np  # for data wrangling 
import folium as folium 
```

```python
# Import csv 
global_partners = pd.read_csv('data/MESA Global Partners - Sheet1.csv')
```

```python
# View csv
global_partners.head()
```

```python
# Data cleaning - get rid of trailing spaces 
global_partners.columns = global_partners.columns.str.strip()
global_partners = global_partners.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
```

```python
# Initiate folium map
m = folium.Map(location=[10, 20], zoom_start=2, 
               tiles="https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}",
    attr="Esri")
```

```python
css = """
<style>
    @import url('https://fonts.googleapis.com/css2?family=Sansita&display=swap');
    .leaflet-popup-content, .leaflet-tooltip {
        font-family: 'Sansita', sans-serif;
        color: #6d0304;
    }
    .leaflet-popup-content a {
        color: #6d0304;
    }
</style>
"""
m.get_root().header.add_child(folium.Element(css))

for _, row in global_partners.iterrows():
    label = row["Abbreviation"] if pd.notna(row["Abbreviation"]) and row["Abbreviation"] else row["Name"]
    popup_html = f'<b>{row["Name"]}</b><br>{row["Region"]}<br><a href="{row["LINK"]}" target="_blank">Visit website</a>'
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=6,
        color='white',
        fill_color='#6d0304',
        fill_opacity=1,
        weight=2,
        popup=folium.Popup(popup_html, max_width=250),
        tooltip=label).add_to(m)
 

m.save("mesa_gp_map.html")
```
